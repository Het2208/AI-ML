```python
import numpy as np

##  Linear Algebra


A = np.array([[2, 1],
              [1, 3]])

# Create vector B
b = np.array([5, 6])

print("Matrix A:\n", A)
print("Vector b:\n", b)

# 1. Inverse
A_inv = np.linalg.inv(A)
print("\nInverse of A:\n", A_inv)

# 2. Determinant
det_A = np.linalg.det(A)
print("\nDeterminant of A:", det_A)

# 3. Eigenvalues & Eigenvectors
eigen_values, eigen_vectors = np.linalg.eig(A)
print("\nEigenvalues:\n", eigen_values)
print("Eigenvectors:\n", eigen_vectors)

# 4. Solve AX = B
X = np.linalg.solve(A, b)
print("\nSolution X for AX = B:\n", X)

# 5. Norm (Vector length)
v = np.array([3, 4])
norm_v = np.linalg.norm(v)
print("\nNorm of vector v:", norm_v)

# 6. Dot Product
dot_product = np.dot(v, v)
print("\nDot product v·v:", dot_product)

# 7. Matrix Multiplication
B = np.array([[1, 2],
              [3, 4]])

result = A @ B
print("\nMatrix Multiplication A @ B:\n", result)
```

```python
arr = np.array([1,2,3,4,5])

print(arr)

print()

# This work slower
for i in range(len(arr)):
    arr[i] += 1
print(arr)

# This run Fast in numpy
# Uses C internally
# 100x faster
print(arr + 1)
```
