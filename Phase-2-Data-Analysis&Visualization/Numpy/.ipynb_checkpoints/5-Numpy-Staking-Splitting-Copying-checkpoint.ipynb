```python
# Stacking in NumPy is the process of joining a sequence of arrays (of the same shape) along a new axis, 
# resulting in a higher-dimensional array.

# Splitting in NumPy is the process of breaking a single array into multiple smaller sub-arrays along a specified axis,
# allowing for easier data manipulation.

# Function	     Purpose
# vstack	     -> stack vertically
# hstack	     -> stack horizontally
# delete	     -> remove rows/columns
# column_stack   ->	combine columns
# row_stack	     -> combine rows
# hsplit         ->	split columns

# 🚀 Real AI/ML Usage
# Operation	         Real Use
# vstack	   -> combine datasets
# hstack	   -> add features
# delete	   -> remove unwanted data
# split	       -> train-test split
# column_stack -> merge feature columns
```

```python
import numpy as np
# Generating Random Arrays
arr1 = np.random.randint(10, size=(2, 2))
arr2 = np.random.randint(10, size=(2, 2))
print(arr1)
print(arr2)
```

```python
# Vertical Stack (vstack)
print(np.vstack((arr1,arr2)))
```

```python
# Horizontal Stack (hstack)
print(np.hstack((arr1,arr2)))
```

```python
# Delete Rows
arr3 = np.delete(arr1 , 1 , 0) 
# Delete index 1 (2nd row)
# axis=0 → rows
print(arr3)
```

```python
#  not imp Column Stack
# ✔ Similar to hstack, but handles 1D arrays better
print(np.column_stack((arr1,arr2)))
```

```python
# not imp Row Stack
print(np.row_stack((arr1,arr2)))
```

```python
arr = np.random.randint(10 , size=(2,10))
print(arr)
```

```python
# Horizontal Split (hsplit)
np.hsplit(arr , 5)
```

```python
# Split at specific indices.
np.hsplit(arr  , (2,4))
```
