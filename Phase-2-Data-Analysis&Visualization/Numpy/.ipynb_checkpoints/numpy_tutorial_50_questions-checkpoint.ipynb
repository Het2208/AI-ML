```python
# Q25. Consider an integer vector Z, which of these expressions are legal?
import numpy as np
Z = np.arange(5)
print(bool(Z**Z))          # ambiguous truth value if array >1 element → but Z**Z is array
# Legal expressions:
print(bool(Z[0]**Z[0]))    # True (0**0 = 1)
print(2 << Z >> 2)         # Legal — bitwise shift
print(Z <- Z)              # Legal — comparison
print(1j * Z)              # Legal — complex multiply
# Z < -Z  is also legal
```

```python
# Q26. What are the result of the following expressions?
import numpy as np
print(np.array(0) / np.array(0))         # nan
print(np.array(0) // np.array(0))        # 0 (integer floor div)
print(np.array([np.nan]).astype(int).astype(float))  # [-9.22e+18] or [nan]
# More readable:
Z = np.linspace(0, 1, 9)
print(Z)
```

```python
# Q27. How to round away from zero a float array?
import numpy as np
Z = np.random.uniform(-10, 10, 10)
print(np.copysign(np.ceil(np.abs(Z)), Z))
```

```python
# Q28. How to find common values between two arrays?
import numpy as np
Z1 = np.random.randint(0, 10, 10)
Z2 = np.random.randint(0, 10, 10)
print(np.intersect1d(Z1, Z2))
```

```python
# Q29. How to ignore all numpy warnings (not recommended)?
import numpy as np
# Method 1: Using context manager (recommended)
with np.errstate(divide='ignore', invalid='ignore'):
    result = np.array([1, 0]) / np.array([0, 0])
print(result)  # [inf nan]

# Method 2: Global suppression (not recommended)
# defaults = np.seterr(all="ignore")
```

```python
# Q30. Is the following expressions true? sqrt(-1) == 1j
import numpy as np
print(np.sqrt(-1) == 1j)           # False (nan)
print(np.emath.sqrt(-1) == 1j)     # True (complex math)
```

```python
# Q31. How to get the dates of yesterday, today and tomorrow?
import numpy as np
yesterday = np.datetime64('today', 'D') - np.timedelta64(1, 'D')
today     = np.datetime64('today', 'D')
tomorrow  = np.datetime64('today', 'D') + np.timedelta64(1, 'D')
print("Yesterday:", yesterday)
print("Today:    ", today)
print("Tomorrow: ", tomorrow)
```

```python
# Q33. How to compute ((A+B)*(-A/2)) in place (without copy)?
import numpy as np
A = np.ones(3) * 1
B = np.ones(3) * 2
C = np.ones(3) * 3
np.add(A, B, out=B)
np.divide(A, 2, out=A)
np.negative(A, out=A)
np.multiply(A, B, out=A)
print(A)
# Simpler example:
A = np.ones(5) * 2
B = np.ones(5) * 3
np.add(A, B, out=B)
np.negative(A, out=A)
np.divide(A, 2, out=A)
np.multiply(B, A, out=A)
print(A)
```

```python
# Q34. Extract the integer part of a random array of positive numbers using 4 different methods.
import numpy as np
Z = np.random.uniform(0, 10, 10)
print(Z - Z % 1)              # Method 1: modulo
print(np.floor(Z))            # Method 2: floor
print(np.ceil(Z) - 1)        # Method 3: ceil minus 1
print(Z.astype(int))         # Method 4: cast
print(np.trunc(Z))           # Method 5: trunc
```

```python
# Q36. Consider a generator function that generates 10 integers and use it to build an array.
import numpy as np
def generate():
    for x in range(10):
        yield x
Z = np.fromiter(generate(), dtype=float, count=-1)
print(Z)
```

```python
# Q40. Consider two random arrays A and B, check if they are equal.
import numpy as np
A = np.random.randint(0, 2, 5)
B = np.random.randint(0, 2, 5)
# Element-wise comparison that handles NaN
equal = np.allclose(A, B)
print(equal)
# Strict equality:
equal_strict = np.array_equal(A, B)
print(equal_strict)
```

```python
# Q41. Make an array immutable (read-only).
import numpy as np
Z = np.zeros(10)
Z.flags.writeable = False
try:
    Z[0] = 1
except ValueError as e:
    print("Error:", e)

# Output: Error: assignment destination is read-only
Z2 = np.ones(10)
print(Z2)
```

```python
# Q42. Consider a random 10×2 matrix representing cartesian coordinates, 
#      convert them to polar coordinates.
import numpy as np
Z = np.random.random((10, 2))
X, Y = Z[:, 0], Z[:, 1]
R = np.sqrt(X**2 + Y**2)
T = np.arctan2(Y, X)
print(R)
print(T)
```

```python
# Q44. Create a structured array with `x` and `y` coordinates covering the [0,1]×[0,1] area.
import numpy as np
Z = np.zeros((5, 5), [('x', float), ('y', float)])
Z['x'], Z['y'] = np.meshgrid(np.linspace(0, 1, 5), np.linspace(0, 1, 5))
print(Z)
```

```python
# Q45. Given two arrays, X and Y, construct the Cauchy matrix C (Cij = 1/(xi - yj)).
import numpy as np
X = np.arange(8)
Y = X + 0.5
C = 1.0 / np.subtract.outer(X, Y)
print(np.linalg.det(C))
```

```python
# Q46. Print the minimum and maximum representable value for each numpy scalar type.
import numpy as np
for dtype in [np.int8, np.int16, np.int32, np.int64]:
    info = np.iinfo(dtype)
    print(f"{dtype.__name__} min={info.min}, max={info.max}")
for dtype in [np.float16, np.float32, np.float64]:
    info = np.finfo(dtype)
    print(f"{dtype.__name__} min={info.min}, max={info.max}")
```

```python
# Q48. How to find the closest value (to a given scalar) in a vector?
import numpy as np
Z = np.arange(100)
v = 3.6
index = (np.abs(Z - v)).argmin()
print(Z[index])   # 4
# Another example:
Z = np.random.uniform(0, 1, 100)
v = 0.5
index = (np.abs(Z - v)).argmin()
print(Z[index])
```

```python
# Q50. Consider a random vector with shape (100,2) representing coordinates,
#      find point by point distances.
import numpy as np

# Show the method using broadcasting:
Z = np.random.random((10, 2))
X, Y = np.atleast_2d(Z[:, 0]), np.atleast_2d(Z[:, 1])
D = np.sqrt((X - X.T)**2 + (Y - Y.T)**2)
print("Distance matrix shape:", D.shape)

# Simpler example — 5×5 grid distances from origin:
grid = np.arange(25).reshape(5, 5)
print(grid)
rows, cols = np.indices((5, 5))
distances = np.sqrt(rows**2 + cols**2)
print("Distances:")
print(distances)
```
