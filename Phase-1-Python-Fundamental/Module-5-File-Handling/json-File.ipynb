```python
import json

people_string = '''
{
  "people": [
    {
      "name": "John Smith",
      "phone": "615-555-7164",
      "emails": ["johnsmith@bogusemail.com", "john.smith@work-place.com"],
      "has_license": false
    },
    {
      "name": "Jane Doe",
      "phone": "560-555-5153",
      "emails": null,
      "has_license": true
    }
  ]
}
'''

data = json.loads(people_string)
# print(type(data)) # <class 'dict'>
# print(data)

for person in data['people']:
    print(person['name'])
```

```python
import json
with open('states.json') as jsonFile:
    data = json.load(jsonFile)
for line in data['states']:
    print(line['name'] , line['abbreviation']) 
```

```python
with open('states.json') as jsonFile:
    data = json.load(jsonFile)
for line in data['states']:
    del line['area_codes']

with open('new-states.json' , 'w') as newJsonFile:
    json.dump(data , newJsonFile)
```
