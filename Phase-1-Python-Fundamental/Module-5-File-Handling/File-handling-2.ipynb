### Practical 1: Count Lines, Words and Characters

**Problem Definition:**  
Write a Python program to read text from a file and count the total number of lines, words, and characters. The result should be written to an output file.


```python
lineCount , wordCount , charCount = 0 , 0  , 0

with open("./File1.txt") as fp:
    file = fp.read()
    word = file.split()
    lineCount = file.count("\n") + 1
    for i in word:
        wordCount += 1
        charCount += len(i) + 1
        
with open("./Output.txt" , "w+") as fp:
    fp.write("lineCount : " + str(lineCount) + "\n")
    fp.write("wordCount : " + str(wordCount) + "\n")
    fp.write("charCount : " + str(charCount) + "\n")
    fp.seek(0,0)
    print(fp.read())
```


### Practical 2: Search a Word in a File

**Problem Definition:**  
Write a Python program to search a given word in a text file and count its occurrences. Store the result in an output file.


```python
def FreqCount(search , Word):
    return Word.count(search)

search = input("Enter a word to search : ")

with open("File1.txt") as fp:
    line = fp.read()
    word = line.split()
    f = FreqCount(search,word)
    with open("FreqCount" , "w+") as fp2:
        fp2.write("Frequency of " + search + " : " + str(f))
        fp2.seek(0,0)
        print(fp2.read())
```


### Practical 3: Student Marks and Average

**Problem Definition:**  
Write a Python program to read roll number, name, and marks of students from a file. Store marks in a list, calculate the average using a user-defined function, and write student details along with average marks to an output file.


```python
def Average(li):
    return sum(li) / len(li)

with open("Student.txt") as fp:
    with open("AverageStudent.txt" , "w+") as fp2:
        for i in fp:
            word = i.split()
            roll = word[0]
            name = word[1]
            mark = word[2:]
            marks = list(map(int , mark))
            avg = Average(marks)
            fp2.write("Roll : "+str(roll) + "| Name : "+name + "| Average : "+str(avg) + "\n")
        fp2.seek(0,0)
        print(fp2.read())
```


### Practical 4: Maximum and Minimum using Tuple

**Problem Definition:**  
Write a Python program to read numbers from a file, store them in a tuple, and find the maximum and minimum values. Write the result to an output file.


```python
with open("score.txt") as fp:
    with open("MinMaxOutput.txt" , "w+") as fp2:
        line = fp.read()
        word = line.split()
        num = tuple(map(int , word))
        Max = max(num)
        Min = min(num)
        fp2.write("Minimum : "+ str(Min) + "\n")
        fp2.write("maximum : "+ str(Max) + "\n")
        fp2.seek(0,0)
        print(fp2.read())
```


### Practical 5: Remove Duplicate Words using Set

**Problem Definition:**  
Write a Python program to read words from a file and remove duplicate words using a set. Store the unique words in an output file.


```python
with open("File1.txt") as fp:
    with open("UniqueOutput.txt" , "w+") as fp2:
        line = fp.read()
        word = line.split()
        words = set(word)
        for i in words:
            fp2.write(i + " ")
        fp2.seek(0,0)
        print(fp2.read())
```


### Practical 6: Word Frequency using Dictionary

**Problem Definition:**  
Write a Python program to count the frequency of each word in a text file using a dictionary and write the result to an output file.


```python
def FreqCount(search , Word):
    return Word.count(search)

with open("File1.txt") as fp:
    with open("dictOutput.txt" , "w+") as fp2:
        line = fp.read()
        word = line.split()
        d = {}
        for i in word:
            d[i] = FreqCount(i , word)
            fp2.write(i + " : " + str(d[i]) + "\n")
        fp2.seek(0,0)
        print(fp2.read())
```


### Practical 7: Student Result and Grade Calculation

**Problem Definition:**  
Write a Python program to read roll number, name, and marks of students from a file. Use a user-defined function to determine grade (A/B/C/Fail) and write the student result to an output file.

**Grade Criteria:**  
- Marks ≥ 75 : Grade A  
- Marks ≥ 60 : Grade B  
- Marks ≥ 40 : Grade C  
- Marks < 40 : Fail


```python
def Average(li):
    return sum(li) / len(li)

def grade(avg):
    if avg >= 75:
        return "A"
    elif avg >= 60:
        return "B"
    elif avg >= 40:
        return "C"
    else:
        return "Fail"
    
with open("Student.txt") as fp:
    with open("gradeStudent.txt" , "w+") as fp2:
        for i in fp:
            word = i.split()
            roll = word[0]
            name = word[1]
            mark = word[2:]
            marks = list(map(int , mark))
            avg = Average(marks)
            g = grade(avg)
            fp2.write("Roll : "+str(roll) + "| Name : "+name + "| Grade : "+g + "\n")
        fp2.seek(0,0)
        print(fp2.read())
```
