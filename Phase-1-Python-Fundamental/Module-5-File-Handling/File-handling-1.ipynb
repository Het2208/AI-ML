### 05) WAP to count the no. of lines, words and characters in a given text file.

```python
lineCount , wordCount , charCount = 0 , 0  , 0
with open("./File1.txt") as fp:
    file = fp.read()
    word = file.split()
    lineCount = file.count("\n") + 1
    for i in word:
        wordCount += 1
        charCount += len(i) + 1

print(f"Number of lines = {lineCount}")
print(f"Number of words = {wordCount}")
print(f"Number of chars = {charCount}")
```

### 09) WAP to get the score of five subjects from the user, store them in a file. Fetch those marks and find the highest score.

### 10) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
def isPrime(n):
    if(n < 2):
        return False
    else:
        for i in range(2 , n):
            if(n%i == 0):
                return False
                break
        else:
            return True


with open("primenumber.txt" , "w+") as fp:
    n = 2
    count = 0
    while(count < 100):
        if(isPrime(n)):
            fp.write(str(n) + "\n")
            count+=1
        n+=1
    fp.seek(0,0)

with open("primenumber.txt") as fp:
    print(fp.read())
```
