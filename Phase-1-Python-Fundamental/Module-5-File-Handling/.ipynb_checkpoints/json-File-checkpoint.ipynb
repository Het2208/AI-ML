```python
import json

people_string = '''
{
  "people": [
    {
      "name": "John Smith",
      "phone": "615-555-7164",
      "emails": ["johnsmith@bogusemail.com", "john.smith@work-place.com"],
      "has_license": false
    },
    {
      "name": "Jane Doe",
      "phone": "560-555-5153",
      "emails": null,
      "has_license": true
    }
  ]
}
'''

data = json.loads(people_string)
# print(type(data)) # <class 'dict'>
# print(data)

for person in data['people']:
    print(person['name'])
```

```python
# for person in data['people']:
#     del person['phone']

newString = json.dumps(data , indent = 2 , sort_keys = True)
print(newString)
```

```python
import json
with open('states.json') as jsonFile:
    data = json.load(jsonFile)
for line in data['states']:
    print(line['name'] , line['abbreviation']) 
```

```python
with open('states.json') as jsonFile:
    data = json.load(jsonFile)
for line in data['states']:
    del line['area_codes']

with open('new-states.json' , 'w') as newJsonFile:
    json.dump(data , newJsonFile)
```

```python
import json
from urllib.request import urlopen

with urlopen("https://finance.yahoo.com/webservice/v1/symbols/allcurrencies/quote?format=json") as response:
    source = response.read()
    
data = json.loads(source)

usd_rates = {}

for item in data['list']['resources']:
    name = item['resource']['fields']['name']
    price = item['resource']['fields']['price']
    usd_rates[name] = price

print(50 * float(usd_rates['USD/EUR']))
```
