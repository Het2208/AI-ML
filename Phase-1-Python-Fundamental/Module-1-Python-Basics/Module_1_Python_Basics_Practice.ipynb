### Q16. Logical Operators — Leap Year

Write a function `is_leap_year(year)` that returns `True` if the year is a leap year.

**Rules:**
- Divisible by 4 → leap year
- EXCEPT divisible by 100 → NOT a leap year
- EXCEPT divisible by 400 → IS a leap year

**Examples:** 2000 ✅, 1900 ❌, 2024 ✅, 2023 ❌

````python
# Q16 — Write is_leap_year(year)

# YOUR CODE HERE
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False
````

### Q20. 🏁 Final Challenge — Mini Calculator

Write a function `calculator(a, op, b)` where:
- `a` and `b` are numbers
- `op` is a string: `"+"`, `"-"`, `"*"`, `"/"`, `"%"`, `"**"`
- Returns the correct result
- If `op` is `"/"` and `b == 0` → return `"Error: Division by zero"`
- If `op` is invalid → return `"Error: Unknown operator"`

**Examples:**
```python
calculator(10, '+', 5)   → 15
calculator(10, '/', 0)   → 'Error: Division by zero'
calculator(2,  '**', 8)  → 256
calculator(10, '?', 5)   → 'Error: Unknown operator'
```

````python
# Q20 — Write the calculator(a, op, b) function

# YOUR CODE HERE
def calculator(a, op, b):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        if b == 0:
            return "Error: Division by zero"
        return a / b
    elif op == "%":
        if b == 0:
            return "Error: Division by zero"
        return a % b
    elif op == "**":
        return a ** b
    else:
        return "Error: Unknown operator"
````

````python
# ── TEST Q20 (do not modify) ──
try:
    check(20,
          calculator(10, '+', 5)  == 15 and
          calculator(10, '-', 3)  == 7  and
          calculator(4,  '*', 5)  == 20 and
          calculator(10, '/', 4)  == 2.5 and
          calculator(10, '/', 0)  == 'Error: Division by zero' and
          calculator(17, '%', 5)  == 2  and
          calculator(2,  '**', 8) == 256 and
          calculator(10, '?', 5)  == 'Error: Unknown operator',
          got=[calculator(10,'+',5), calculator(10,'/',0)],
          expected=[15, 'Error: Division by zero'])
except Exception as e:
    print(f'❌ Q20 FAILED — Error: {e}')
````
