### 01) WAP to find their union, intersection, and difference of the given two sets.
**Sample Input:**
```
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

```
**Sample Output:**
```
Union: {1, 2, 3, 4, 5, 6}
Intersection: {3, 4}
Difference (A-B): {1, 2}

```

````python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print(f"Union : {A|B}")
print(f"Intersection : {A&B}")
print(f"Diffrence (A-B) : {A-B}")
````

````python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
U = set({})
I = set({})
D = set({})
#Union
for x in A:
    for i in B:
        if x not in B or i not in A:
            U.add(x)
        if i not in A or x not in B:
            U.add(i)
#Intersection
for x in A:
    if x in B:
        I.add(x)

#Diffrence (A-B) 
for x in A:
    for i in B:
        if x not in B:
            D.add(x)
        
print(f"Union : {U}")
print(f"Intersection : {I}")
print(f"Diffrence (A-B) : {D}")
````

### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}
print(Dict1 | Dict2)
````

````python
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}
res = Dict1.copy()
for x,y in Dict2.items():
    if x in res:
        res[x] += y
    else:
        res[x] = y
print(res)
````

### 06) WAP to Sort a dictionary in ascending order of its values.
Hint: in sorted(), use udf for key parameter


**Sample Input:**
```
{'Math': 70, 'Physics': 85, 'Chemistry': 60}

```
**Sample Output:**
```
{'Chemistry': 60, 'Math': 70, 'Physics': 85}
```

````python
def sortByValue(item):
    return item[1]

d = {'Math': 70, 'Physics': 85, 'Chemistry': 60}

sorted_dict = dict(sorted(d.items(), key=sortByValue))
print(sorted_dict)
````

### 07) WAP to find the keys having maximum and minimum values in a dictionary.

**Sample Input:**
```
{'A': 50, 'B': 90, 'C': 30}
```
**Sample Output:**
```
Max Value Key: B
Min Value Key: C

```

````python
d = {'A': 50, 'B': 90, 'C': 30}
print(f"Max Value Key : {max(d, key=d.get)}")
print(f"Min Value Key : {min(d, key=d.get)}")
````

````python
d = {'A': 50, 'B': 90, 'C': 30}
keys = list(d.keys())
Min , Max = d[keys[0]] , d[keys[0]]
Min_Key = Max_Key = keys[0]
for x,y in d.items():
    if y > Max:
        Max = y
        Max_Key = x
    if y < Min:
        Min = y
        Min_Key = x
print(f"Max Value Key : {Max_Key}")
print(f"Min Value Key : {Min_Key}")    
````

dict1 = {'a': 5, 'c': 8, 'e': 2}

Key Not Found

### 15) WAP to create a dictionary using dictionary comprehension from two lists containing student names and marks. Include only those students in the dictionary whose marks are greater than or equal to 75.

**Sample Input:**
```
students = ['Amit', 'Riya', 'Neha', 'Karan']
marks = [78, 72, 85, 60]
```
**Sample Output:**
```
{'Amit': 78, 'Neha': 85}
```

````python
students = ['Amit', 'Riya', 'Neha', 'Karan']
marks = [78, 72, 85, 60]

res = {}

for y, x in zip(students, marks):
    if x > 75:
        res[y] = x

print(res)
````
