### Q9. Word Frequency Counter

Write a function `word_frequency(sentence)` that returns a dict with each word as key and its count as value. Case-insensitive (convert to lowercase).

**Examples:**
```python
word_frequency('the cat sat on the mat the cat')
→ {'the': 3, 'cat': 2, 'sat': 1, 'on': 1, 'mat': 1}

word_frequency('Hello hello HELLO')
→ {'hello': 3}
```

````python
# Q9 — Write word_frequency(sentence)

# YOUR CODE HERE
def word_frequency(sentence):
    sentence = sentence.lower()
    word = sentence.split()
    d = {}
    for i in word:
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
    return d
````

```python

only_exam1 → {'English'}

```

````python
# Q10 — Merge and compare dictionaries
exam1 = {'Math':85, 'Science':90, 'English':78}
exam2 = {'Math':92, 'Science':88, 'History':75}

# YOUR CODE HERE
merged     = {}
common     = {}
only_exam1 = set()
only_exam2 = set()
for i,j in (exam1.items()):
    if i in exam2:
        merged[i] = max(j,exam2[i])
        common[i] = j
    else:
        merged[i] = j
        only_exam1.add(i)
for i,j in (exam2.items()):
    if i not in exam1:
        merged[i] = j
        only_exam2.add(i)

print(merged)
print(common)
print(only_exam1)
print(only_exam2)
````
