## 1. Basic List Comprehensions

A **list comprehension** provides a concise way to create lists.

**Syntax:**
```python
[expression for item in iterable]
```

**Example (without comprehension):**
```python
squares = []
for i in range(10):
    squares.append(i ** 2)
```

**Same thing with comprehension:**
```python
squares = [i ** 2 for i in range(10)]
```

### ✏️ Practice Questions

**Q1.** Use a list comprehension to create a list of the first 10 natural numbers (1 to 10).

**✅ Answer Q1:**

````python
numbers = [i for i in range(1, 11)]
print(numbers)
````

**Q2.** Use a list comprehension to generate a list of cubes of numbers from 1 to 8.

**✅ Answer Q2:**

````python
cubes = [i ** 3 for i in range(1, 9)]
print(cubes)
````

**Q3.** Given the string `"hello world"`, use a list comprehension to get a list of all characters (including spaces).

**✅ Answer Q3:**

````python
chars = [ch for ch in "hello world"]
print(chars)
````

## 2. Comprehension with Condition (Filtering)

You can add an **`if` condition** at the end to filter items.

**Syntax:**
```python
[expression for item in iterable if condition]
```

**Example:** Get only even numbers from 0–19
```python
evens = [i for i in range(20) if i % 2 == 0]
```

### ✏️ Practice Questions

**Q1.** Use a list comprehension to get all odd numbers between 1 and 20.

**✅ Answer Q1:**

````python
odds = [i for i in range(1, 21) if i % 2 != 0]
print(odds)
````

**Q2.** From the list `words = ["apple", "ant", "banana", "avocado", "blueberry", "apricot"]`, use a list comprehension to get only words that start with `"a"`.

**✅ Answer Q2:**

````python
words = ["apple", "ant", "banana", "avocado", "blueberry", "apricot"]
a_words = [w for w in words if w.startswith("a")]
print(a_words)
````

**Q3.** From the list `nums = [5, -3, 0, 8, -1, 12, -7, 4]`, extract only positive numbers using a list comprehension.

**✅ Answer Q3:**

````python
nums = [5, -3, 0, 8, -1, 12, -7, 4]
positives = [n for n in nums if n > 0]
print(positives)
````

## 3. Comprehension with Multiple Conditions

You can chain **multiple `if` conditions** to apply more than one filter.

**Syntax:**
```python
[expression for item in iterable if condition1 if condition2]
# OR equivalently:
[expression for item in iterable if condition1 and condition2]
```

**Example:** Numbers divisible by both 2 and 3
```python
result = [i for i in range(50) if i % 2 == 0 if i % 3 == 0]
```

### ✏️ Practice Questions

**Q1.** Use a list comprehension to get all numbers from 1 to 100 that are divisible by both 5 and 7.

**✅ Answer Q1:**

````python
result = [i for i in range(1, 101) if i % 5 == 0 if i % 7 == 0]
print(result)
````

**Q2.** From the list `words = ["sky", "sun", "star", "moon", "storm", "sea"]`, use a comprehension to get words that start with `"s"` AND have more than 3 characters.

**✅ Answer Q2:**

````python
words = ["sky", "sun", "star", "moon", "storm", "sea"]
filtered = [w for w in words if w.startswith("s") if len(w) > 3]
print(filtered)
````

**Q3.** Generate all numbers from 1 to 200 that are even AND divisible by 3 AND greater than 100.

**✅ Answer Q3:**

````python
result = [i for i in range(1, 201) if i % 2 == 0 if i % 3 == 0 if i > 100]
print(result)
````

## 4. Multiple List Comprehension (Nested Loops)

You can use **multiple `for` clauses** in a single comprehension — like nested loops.

**Syntax:**
```python
[expression for item1 in iterable1 for item2 in iterable2]
```

**Example:** Cartesian product of two lists
```python
pairs = [(x, y) for x in [1, 2, 3] for y in ['a', 'b']]
# [(1,'a'), (1,'b'), (2,'a'), (2,'b'), (3,'a'), (3,'b')]
```

### ✏️ Practice Questions

**Q1.** Use a list comprehension with two `for` clauses to generate all pairs `(i, j)` where `i` is from `[1, 2, 3]` and `j` is from `[10, 20]`.

**✅ Answer Q1:**

````python
pairs = [(i, j) for i in [1, 2, 3] for j in [10, 20]]
print(pairs)
````

**Q2.** Use a list comprehension to generate all pairs `(x, y)` where `x` is from `range(1, 4)` and `y` is from `range(1, 4)`, but only where `x != y`.

**✅ Answer Q2:**

````python
pairs = [(x, y) for x in range(1, 4) for y in range(1, 4) if x != y]
print(pairs)
````

**Q3.** Given `suits = ['Hearts', 'Diamonds']` and `ranks = ['Ace', 'King', 'Queen']`, generate a list of all card combinations as strings like `'Ace of Hearts'`.

**✅ Answer Q3:**

````python
suits = ['Hearts', 'Diamonds']
ranks = ['Ace', 'King', 'Queen']
cards = [f"{rank} of {suit}" for suit in suits for rank in ranks]
print(cards)
````

## 5. If/Else in a Comprehension

You can use a **ternary (inline) if/else** in the *expression* part to transform values conditionally.

**Syntax:**
```python
[value_if_true if condition else value_if_false for item in iterable]
```

> ⚠️ Note: `if/else` for transformation goes **before** the `for` keyword. Filtering `if` goes **after**.

**Example:** Label numbers as `"even"` or `"odd"`
```python
labels = ["even" if i % 2 == 0 else "odd" for i in range(6)]
# ['even', 'odd', 'even', 'odd', 'even', 'odd']
```

### ✏️ Practice Questions

**Q1.** For numbers 1 to 10, use a comprehension to produce `"fizz"` if the number is divisible by 3, otherwise the number itself.

**✅ Answer Q1:**

````python
result = ["fizz" if i % 3 == 0 else i for i in range(1, 11)]
print(result)
````

**Q2.** From `nums = [4, -2, 7, -9, 0, 3]`, use a comprehension to create a list where negatives become `0` and positives stay as-is (keep `0` as `0`).

**✅ Answer Q2:**

````python
nums = [4, -2, 7, -9, 0, 3]
clipped = [n if n > 0 else 0 for n in nums]
print(clipped)
````

**Q3.** For each word in `["cat", "elephant", "dog", "hippopotamus", "ant"]`, produce `"long"` if the word has more than 4 characters, else `"short"`.

**✅ Answer Q3:**

````python
animals = ["cat", "elephant", "dog", "hippopotamus", "ant"]
lengths = ["long" if len(a) > 4 else "short" for a in animals]
print(lengths)
````

## 6. Nested List Comprehension

A **nested list comprehension** is when the expression itself is another comprehension — useful for 2D structures like matrices.

**Syntax:**
```python
[[expression for col in cols] for row in rows]
```

**Example:** Create a 3×3 zero matrix
```python
matrix = [[0 for _ in range(3)] for _ in range(3)]
# [[0, 0, 0], [0, 0, 0], [0, 0, 0]]
```

### ✏️ Practice Questions

**Q1.** Use a nested list comprehension to create a 4×4 multiplication table (each inner list is a row).

**✅ Answer Q1:**

````python
table = [[i * j for j in range(1, 5)] for i in range(1, 5)]
for row in table:
    print(row)
````

**Q2.** Given `matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]`, use a nested list comprehension to **transpose** it (rows become columns).

**✅ Answer Q2:**

````python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
transposed = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
for row in transposed:
    print(row)
````

**Q3.** Use a nested list comprehension to **flatten** the list `nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]` into a single list.

**✅ Answer Q3:**

````python
nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
flat = [item for sublist in nested for item in sublist]
print(flat)
````

## 7. Transformation in Comprehension

The expression part of a comprehension can apply any **function or transformation** to each element.

**Example:** Converting strings to uppercase
```python
names = ["alice", "bob", "charlie"]
upper = [name.upper() for name in names]
# ['ALICE', 'BOB', 'CHARLIE']
```

You can even call your own functions inside a comprehension!

### ✏️ Practice Questions

**Q1.** Given `temps_c = [0, 20, 37, 100]` (Celsius), use a comprehension to convert each to Fahrenheit. Formula: `F = C * 9/5 + 32`

**✅ Answer Q1:**

````python
temps_c = [0, 20, 37, 100]
temps_f = [c * 9/5 + 32 for c in temps_c]
print(temps_f)
````

**Q2.** From `sentences = ["  hello world  ", "  python rocks  ", "  keep coding  "]`, use a comprehension to strip whitespace and capitalize each sentence.

**✅ Answer Q2:**

````python
sentences = ["  hello world  ", "  python rocks  ", "  keep coding  "]
cleaned = [s.strip().capitalize() for s in sentences]
print(cleaned)
````

**Q3.** Write a function `double_if_even(n)` that returns `n * 2` if even, else returns `n`. Then use a list comprehension with this function over `range(1, 8)`.

**✅ Answer Q3:**

````python
def double_if_even(n):
    return n * 2 if n % 2 == 0 else n

result = [double_if_even(i) for i in range(1, 8)]
print(result)
````

## 8. Dictionary Comprehension

Just like list comprehensions, you can build **dictionaries** concisely.

**Syntax:**
```python
{key_expr: value_expr for item in iterable}
{key_expr: value_expr for item in iterable if condition}
```

**Example:** Map numbers to their squares
```python
sq = {i: i**2 for i in range(1, 6)}
# {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

### ✏️ Practice Questions

**Q1.** Use a dictionary comprehension to map each word in `["apple", "banana", "cherry"]` to its length.

**✅ Answer Q1:**

````python
words = ["apple", "banana", "cherry"]
word_lengths = {w: len(w) for w in words}
print(word_lengths)
````

**Q2.** Given `students = {"Alice": 85, "Bob": 72, "Charlie": 91, "Diana": 67}`, use a dict comprehension to keep only students who scored above 75.

**✅ Answer Q2:**

````python
students = {"Alice": 85, "Bob": 72, "Charlie": 91, "Diana": 67}
top_students = {name: score for name, score in students.items() if score > 75}
print(top_students)
````

**Q3.** Use a dictionary comprehension to **swap keys and values** of `d = {"a": 1, "b": 2, "c": 3}`.

**✅ Answer Q3:**

````python
d = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in d.items()}
print(inverted)
````

## 9. Set Comprehension

**Set comprehensions** work just like list comprehensions but produce a **set** (unordered, no duplicates).

**Syntax:**
```python
{expression for item in iterable}
```

**Example:** Get unique first letters from a list of names
```python
names = ["Alice", "Bob", "Anna", "Brian", "Charlie"]
initials = {name[0] for name in names}
# {'A', 'B', 'C'}
```

### ✏️ Practice Questions

**Q1.** Use a set comprehension to get all unique vowels in the string `"comprehension is powerful"`.

**✅ Answer Q1:**

````python
text = "comprehension is powerful"
vowels = {ch for ch in text if ch in "aeiou"}
print(vowels)
````

**Q2.** From `nums = [1, 2, 2, 3, 4, 4, 5, 5, 5]`, use a set comprehension to get a set of squares of unique values.

**✅ Answer Q2:**

````python
nums = [1, 2, 2, 3, 4, 4, 5, 5, 5]
squares_set = {n ** 2 for n in nums}
print(squares_set)
````

**Q3.** Use a set comprehension to find the set of lengths of words in `phrase = "to be or not to be that is the question"`.

**✅ Answer Q3:**

````python
phrase = "to be or not to be that is the question"
word_lengths = {len(word) for word in phrase.split()}
print(word_lengths)
````

## 10. Generator Comprehension (Generator Expressions)

A **generator expression** looks like a list comprehension but uses **parentheses `()`** instead of brackets. It returns a **generator object** — values are computed **lazily** (one at a time), saving memory.

**Syntax:**
```python
(expression for item in iterable)
(expression for item in iterable if condition)
```

**Example:**
```python
gen = (i ** 2 for i in range(1000000))  # Uses very little memory!
print(next(gen))   # 0
print(next(gen))   # 1
```

> 💡 Use generators when you only need to **iterate once** or are working with **large datasets**.

### ✏️ Practice Questions

**Q1.** Create a generator expression for squares of numbers 1–10. Use `next()` to print the first 3 values, then use `sum()` to get the total of all remaining values.

**✅ Answer Q1:**

````python
gen = (i ** 2 for i in range(1, 11))
print(next(gen))   # 1
print(next(gen))   # 4
print(next(gen))   # 9
print(sum(gen))    # Sum of remaining: 16+25+36+49+64+81+100 = 371
````

**Q2.** Use a generator expression inside `sum()` to calculate the sum of all even numbers from 1 to 1000 (without building a list).

**✅ Answer Q2:**

````python
total = sum(i for i in range(1, 1001) if i % 2 == 0)
print(total)
# Output: 250500
````

**Q3.** Use a generator expression inside `any()` to check if any word in `["cat", "elephant", "dog", "hippopotamus"]` has more than 8 characters. Then use `all()` to check if all words have more than 2 characters.

**✅ Answer Q3:**

````python
animals = ["cat", "elephant", "dog", "hippopotamus"]

has_long = any(len(a) > 8 for a in animals)
all_above_2 = all(len(a) > 2 for a in animals)

print("Any word longer than 8 chars?", has_long)    # True
print("All words longer than 2 chars?", all_above_2) # True
````
