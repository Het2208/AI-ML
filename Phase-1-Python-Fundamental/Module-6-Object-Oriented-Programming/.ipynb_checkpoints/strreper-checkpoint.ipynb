```python
# When printing an object, Python checks for __str__ first. If it is not defined, it falls back to the __repr__ method
# If both are missing, Python defaults to showing the object type and its memory address 
# The speaker recommends implementing at least the __repr__ method in your classes to ensure both debugging utility and
# a fallback for standard printing

# __str__: Designed for providing a human-readable string representation of an object. It is used when you explicitly call 
# print(), str(), or use formatting functions on an object

# __repr__: Designed for developers to aid in debugging. It provides a more technical, unambiguous representation of an object.
#     This method is invoked when you inspect an object directly in the debugger or the interactive shell

class Employee:
    def __init__(self , name , age):
        self.name = name 
        self.age = age

    def __str__(self):
        return f"Name : {self.name} || Age : {self.age}"

    def __repr__(self):
        return "This is __repr__ method"
        
emp1 = Employee('het' , 19)
print(emp1) 

# without defining __str__ it will print -> <__main__.Employee object at 0x000002671ADE17F0>
# with __str__ method it will print -> Name : het || Age : 19
# without __str__ and with __repr__ it will print -> This is __repr__ method

# both missing -> object 
# __str__ -> str
# __repr__ -> repr
# both -> str
# degugging -> repr not str
```
