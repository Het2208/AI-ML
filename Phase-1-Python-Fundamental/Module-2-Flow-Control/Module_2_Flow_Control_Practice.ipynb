````python
# Q7 — Write digit_sum(n) using a for loop

# YOUR CODE HERE
def digit_sum(n):
    Sum = 0
    while n != 0:
        Sum += n % 10
        n //= 10
    return Sum
````

### Q11. Reverse a Number

Write a function `reverse_num(n)` that reverses the digits of a number using a `while` loop.
**Do NOT use string slicing.**  Use `%` and `//` only.

**Examples:**
```python
reverse_num(1234)   → 4321
reverse_num(100)    → 1
reverse_num(7)      → 7
```

````python
# Q11 — Write reverse_num(n) using a while loop (no string slicing!)

# YOUR CODE HERE
def reverse_num(n):
    rem = 0
    rev = 0
    while n != 0:
        rem = n % 10
        rev = rev * 10 + rem
        n //= 10
    return rev
````

### Q14. Collatz Conjecture

Write a function `collatz(n)` that applies the Collatz sequence and returns how many steps it takes to reach 1.
- If `n` is even → `n = n // 2`
- If `n` is odd  → `n = 3 * n + 1`
- Count each step until n == 1

**Examples:**
```python
collatz(6)    → 8
collatz(1)    → 0
collatz(27)   → 111
```

````python
# Q14 — Write collatz(n)

# YOUR CODE HERE
def collatz(n):
    step = 0
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
        step += 1
    return step
````

### Q15. Fibonacci Sequence

Write a function `fibonacci(n)` that returns a list of the first `n` Fibonacci numbers using a `while` loop.
The sequence starts: 0, 1, 1, 2, 3, 5, 8 ...

**Examples:**
```python
fibonacci(1)   → [0]
fibonacci(5)   → [0, 1, 1, 2, 3]
fibonacci(8)   → [0, 1, 1, 2, 3, 5, 8, 13]
```

````python
# Q15 — Write fibonacci(n) using a while loop

# YOUR CODE HERE
def fibonacci(n):
    res = []
    res.append(0)
    n1 = 0
    n2 = 1
    i = 1
    while i != n:
        n3 = n1 + n2
        res.append(n2)
        n1 = n2
        n2 = n3
        i += 1
    return res
````

```python

```

````python
# Q16 — Write first_prime_after(n) using while + break

# YOUR CODE HERE
def first_prime_after(n):
    num = n + 1
    while True:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            return num
        num += 1
````

````python
# Q18 — Write first_duplicate(lst) using break

# YOUR CODE HERE
def first_duplicate(lst):
    seen = []
    for i in lst:
        if i in seen:
            return i
        seen.append(i)
    return None
````

### Q19. Process Valid Scores

Write a function `process_scores(scores)` that:
- **Skips** invalid scores (< 0 or > 100) using `continue`
- **Stops** processing if score is exactly `0` using `break`
- Returns a dict: `{'valid': [...], 'average': float, 'stopped_early': bool}`

**Example:**
```python
process_scores([80, 90, -5, 70, 0, 60])
→ {'valid': [80, 90, 70], 'average': 80.0, 'stopped_early': True}

process_scores([75, 88, 200, 60])
→ {'valid': [75, 88, 60], 'average': 74.33..., 'stopped_early': False}
```

````python
# Q19 — Write process_scores(scores) using break and continue

# YOUR CODE HERE
def process_scores(scores):
    valid = []
    se = False
    for i in scores:
        if i < 0 or i > 100:
            continue
        elif i == 0:
            se = True
            break
        else:
            valid.append(i)
    return {
        'valid': valid,
        'average': sum(valid) / len(valid) ,
        'stopped_early': se
    }
````

Write a function `guessing_game(secret, guesses)` that simulates a number guessing game.
- Loop through the `guesses` list
- If guess == secret → return `{'found': True, 'attempts': count, 'message': 'Found in N attempts!'}`
- If guess < secret  → hint is `'Too low'`
- If guess > secret  → hint is `'Too high'`
- If secret not found after all guesses → return `{'found': False, 'attempts': len(guesses), 'message': 'Not found!'}`

**Examples:**
```python
guessing_game(42, [10, 60, 42])
→ {'found': True, 'attempts': 3, 'message': 'Found in 3 attempts!'}

guessing_game(42, [1, 2, 3])
→ {'found': False, 'attempts': 3, 'message': 'Not found!'}
```

````python
def guessing_game(secret, guesses):
    attempts = 0

    for guess in guesses:
        attempts += 1

        if guess == secret:
            return {
                'found': True,
                'attempts': attempts,
                'message': f'Found in {attempts} attempts!'
            }
        elif guess < secret:
            hint = 'Too low'
        else:
            hint = 'Too high'

    return {
        'found': False,
        'attempts': len(guesses),
        'message': 'Not found!'
    }
````
