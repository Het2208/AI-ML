### 07) WAP to print sum of series 1 – 2 + 3 – 4 + 5 – 6 + 7 ... n.

```python
n = int(input("Enter a number : "))
Sum = 0
for i in range(1,n+1):
    if i % 2 == 0:
        Sum -= i
    else:
        Sum += i

print(f"sum = {Sum}")
```

### 10) WAP to print GCD of given two numbers.

```python
n1 = int(input("Enter a number : "))
n2 = int(input("Enter a number : "))
lim = min(n1,n2)
GCD = 0
for i in range(1,lim):
    if n1%i == 0 and n2%i == 0:
        GCD = i
print(f"GCD = {GCD}")
    
```

### 12) WAP to find whether the given number is Prime or not.

```python
n = int(input("Enter a number : "))
if n < 2:
    print(f"{n} is not prime number")
else:
    for i in range(2,n):
        if n % i == 0:
            print(f"{n} is not prime number")
            break    
    else:
         print(f"{n} is prime number")
```

### 13) WAP to print sum of digits of given number.

```python
n = int(input("Enter a number : "))

temp = n
Sum = 0
rem = 0

while temp != 0:
    rem = temp%10
    Sum += rem
    temp //= 10

print(f"Sum all individual degit = {Sum}")
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input("Enter a number : "))

temp = n
rev = 0
rem = 0

while temp != 0:
    rem = temp%10
    rev = (rev*10) + rem
    temp //= 10

if n == rev:
    print(f"{n} is Palindrome")
else:
    print(f"{n} is not Palindrome")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
n = int(input("Enter a number to check Armstrong : "))

temp  = n
count = 0

while temp != 0:
    count += 1
    temp //= 10

temp = n
arm=0
while temp != 0:
    rem = temp%10
    arm += pow(rem,count)
    temp //= 10

if n == arm:
    print(f"{n} is Armstrong Number")
else:
    print(f"{n} is Not Armstrong Number")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
n = int(input("Enter a number : "))


for i in range(1, n+1):
    Sum = 0
    for j in range(1,i):
        if i % j == 0:
            Sum += j
    if Sum == i:
        print(i,end=" ")
        
```
