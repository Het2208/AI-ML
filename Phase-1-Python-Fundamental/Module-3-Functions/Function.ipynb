### 01) Write a function to calculate BMI given mass and height. (BMI = mass/h**2)

```python
def bmi(mass , height):
    return mass/height**2

height = int(input("Enter Height : "))
mass = int(input("Enter mass : "))
print(f"BMI = {bmi(height,mass)}")
```

### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def isPrime(n):
    flag = 1
    if(n<2):
        flag = 0
    else:
        for i in range(2,n):
            if(n%i) == 0:
                flag = 0
                break
        else:
            flag = 1
    return flag
    
n = int(input("Enter a number : "))
print(isPrime(n))
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
def PrimeList(n1,n2):
    primeList = []
    for i in range(n1,n2+1):
        if(isPrime(i) == 1):
            primeList.append(i)
    return primeList
n1 = int(input("Enter n1 : "))
n2 = int(input("Enter n2 : "))

print(PrimeList(n1,n2))
```

### 05) Write a function that returns True if the given string is Palindrome or False otherwise.

```python
def stringPalindrome(Str):
    revStr = Str[::-1]
    if(revStr == Str):
        return True
    else:
        return False

Str = input("Enter a String : ").lower()
print(stringPalindrome(Str))
```

### 07) Write a function to calculate the sum of the first element of each tuples inside the list.

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
        
n = int(input("Enter a number : "))
(fibonacci(n))
```

### 10) Write a function to get the sum of the scores ending with zero.
#### Example : scores = [200, 456, 300, 100, 234, 678] 
#### Ans = 200 + 300 + 100 = 600

```python
def SumScoreEndZero(s):
    Sum = 0
    for i in s:
        if(i % 10) == 0:
            Sum += i
    return Sum
    
scores = [200, 456, 300, 100, 234, 678]
print(SumScoreEndZero(scores))
```

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
def invertKeyValue(d):
    res = {}
    for x,y in d.items():
        res[y] = x
    return res
d1 = {'a': 10, 'b':20, 'c':30, 'd':40}
print(invertKeyValue(d1))
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
def UppLow(str):
    uCount = 0
    lCount = 0
    for i in str:
        if(i.islower()):
            lCount += 1
        else:
            uCount += 1
    return uCount , lCount

str = input("Enter a string : ")
print(f"no_upper {UppLow(str)[0]}")
print(f"no_lower {UppLow(str)[1]}")
```

### 16) Write udfs to call the functions with following types of arguments:
1. Positional Arguments
2. Keyword Arguments
3. Default Arguments
4. Variable Legngth Positional(*args) & variable length Keyword Arguments (**kwargs)
5. Keyword-Only & Positional Only Arguments

```python
def add(a, b):
    return a + b

def add_keyword(a, b):
    return a + b

def add_default(a=2, b=9):
    return a + b

def add_args(*numbers):
    return sum(numbers)

def show_kwargs(**data):
    return data

def keyword_only(*, a, b):
    return a + b

def positional_only(a, b, /):
    return a + b

print(f"Positional Arguments : {add(2, 2)}")
print(f"Keyword Arguments : {add_keyword(b=5, a=2)}")
print(f"Default Arguments : {add_default()}")
print(f"*args (Variable Positional) : {add_args(1, 2, 3, 4)}")
print(f"**kwargs (Variable Keyword) : {show_kwargs(name='Het', age=21)}")
print(f"Keyword-Only Arguments : {keyword_only(a=3, b=7)}")
print(f"Positional-Only Arguments : {positional_only(4, 6)}")
```
