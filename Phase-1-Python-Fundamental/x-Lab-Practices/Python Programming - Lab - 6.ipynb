### 01) WAP to find sum of all the elements in a List.

```python
n = int(input("Enter a size of list"))
l1 = []
for i in range(n):
    data = int(input("Enter a  element : "))
    l1.append(data)
Sum = 0
for i in l1:
    Sum += i
print(f"Sum = {Sum}")
```

### 02) WAP to find largest element in a List.

```python
n = int(input("Enter a size of list"))
l1 = []
for i in range(n):
    data = int(input("Enter a  element : "))
    l1.append(data)
Max = l1[0]
for i in l1:
    if i > Max:
        Max = i
print(f"Largest Element = {Max}")
```

### 04) WAP to interchange first and last elements in a list.

```python
n = int(input("Enter a size of list"))
l1 = []
for i in range(n):
    data = int(input("Enter a  element : "))
    l1.append(data)
print(f"Before Interchange :{l1}")
l1[0],l1[-1] = l1[-1],l1[0]
print(f"After Interchange  :{l1}")
```

### 05) WAP to split the List into two parts and append the first part to the end.

```python
n = int(input("Enter a size of list"))
middle = n//2
l1 = []
for i in range(n):
    data = int(input("Enter a  element : "))
    l1.append(data)

l1 = l1[middle:] + l1[:middle:]

print(l1)
```

### 06) WAP to interchange the elements on two positions entered by a user.

```python
n = int(input("Enter a size of list"))
l1 = []
for i in range(n):
    data = int(input("Enter a  element : "))
    l1.append(data)

pos1 = int(input("Enter a Position 1 : "))
pos2 = int(input("Enter a Position 2 : "))

print(f"Before interchange : {l1}")

l1[pos1] , l1[pos2] = l1[pos2] , l1[pos1]

print(f"After interchange : {l1}")
```

### 08) WAP to print even numbers in a list.

### 09) WAP to count unique items in a list.

```python
n = int(input("Enter a size of list"))
l1 = []
for i in range(n):
    data = int(input("Enter a  element : "))
    l1.append(data)
count = 0
l2 = []
for i in l1:
    if i not in l2:
        count += 1
        l2.append(i)
print(count)
```

### 12) WAP to count occurrences of an element in a list.

```python
n = int(input("Enter a size of list"))
l1 = []
for i in range(n):
    data = int(input("Enter a  element : "))
    l1.append(data)
l2 = []
for i in l1:
    count = l1.count(i)
    l2.append(i)
    if l2.count(i) == 1:
        print(f"{i} => {count}")
```

### 14) WAP to extract elements with frequency greater than K.

```python
n = int(input("Enter a size of list"))
l1 = []
for i in range(n):
    data = int(input("Enter a  element : "))
    l1.append(data)
k = int(input("Enter K : "))
l2 = []
for i in l1:
    count = l1.count(i)
    l2.append(i)
    if l2.count(i) == 1:
        if count > k:
            print(f"{i} => {count}")
```

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
n = int(input("Enter a size of list"))
l1 = []
for i in range(n):
    data = input("Enter a  element : ").lower()
    l1.append(data)
l2 = []
for i in l1:
    if i.startswith("b"):
        l2.append(i)

print(l2)
```

### 17) WAP to create a list of common elements from given two lists.

```python
n = int(input("Enter a size of list"))
l1 = []
for i in range(n):
    data = input("Enter a  element : ")
    l1.append(data)
n = int(input("Enter a size of list"))
l2 = []
for i in range(n):
    data = input("Enter a  element : ")
    l2.append(data)

l3 = []
for i in l1:
    if i in l2:
        l3.append(i)
print(l3)
```
