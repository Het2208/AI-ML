### 01) WAP to find sum of tuple elements.
**Sample Input:**
```
T = (10, 20, 30, 40)

```
**Sample Output:**
```
Sum of tuple elements = 100
```

````python
n = int(input("Enter a number of element : "))
l = [] 
Sum = 0
for i in range(n):
    ele = int(input(f"Enter a element {i+1} : "))
    l.append(ele)
tup = tuple(l)
for i in tup:
    Sum += i
print(f"Sum of tuple elements = {Sum}")
````

### 03) WAP to find tuples which have all elements divisible by K from a list of tuples.
**Sample Input:**
```
tuple_list = [(2, 4, 6), (3, 6, 9), (4, 8, 12), (5, 10, 15)]
K = 2

```
**Sample Output:**
```
Tuples divisible by 2 : [(2, 4, 6), (4, 8, 12)]
```

````python
tuple_list = [(2, 4, 6), (3, 6, 9), (4, 8, 12), (5, 10, 15)]
k = int(input("Enter a K :"))
res = []
for i in tuple_list:
    for x in i:
        if x % k != 0:
            break
    else:
        res.append(i)
print(f"Tuples divisible by {k} : {res}")
````

### 10) WAP to test if the given tuple is distinct or not.
**Sample Input:**
```
tuple1 = (1, 2, 3, 4)
tuple2 = (1, 2, 2, 3)
```
**Sample Output:**
```
(1, 2, 3, 4) is distinct? -> True
(1, 2, 2, 3) is distinct? -> False
```

````python
tuple1 = (1, 2, 3, 4)
tuple2 = (1, 2, 2, 3)
res = []
for i in tuple1:
    if i in res:
        print(f"{tuple1}  is distinct? -> False")
        break
    res.append(i)
else:
    print(f"{tuple1}  is distinct? -> True")
res.clear()
for i in tuple2:
    if i in res:
        print(f"{tuple2}  is distinct? -> False")
        break
    res.append(i)
else:
    print(f"{tuple2}  is distinct? -> True")       
````

### 14) WAP to create (index, value) pairs for prime numbers.

**Sample Input:**
```
T = (4, 7, 9, 11, 15)
```
**Sample Output:**
```
((1, 7), (3, 11))
```

````python
def prime(n):
    flag = True
    if n < 2:
        flag = False
    for i in range(2 , n+1):
        if n%i == 0:
            flag = False
            break
    else:
        flag = True
    return flag

T = (4, 7, 9, 11, 15)
res = tuple([(i , T[i]) for i in range(len(T)) if (prime(i) == True)])

print(res)
````

### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])

rows = len(T)
cols = len(T[0])

for j in range(cols):
    col_sum = 0
    for i in range(rows):
        col_sum += T[i][j]
    print(col_sum, end=" ")
````
