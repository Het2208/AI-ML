### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
str = input("Enter a String : ")
uppercount , lowercount , specialcount , digitcount = 0, 0 , 0 , 0
for i in str:
    if i.islower():
        lowercount += 1
    elif i.isupper():
        uppercount += 1
    elif i.isdigit():
        digitcount += 1
    else :
        specialcount += 1

print(f"Uppercase : {uppercount} Lowercount : {lowercount} Digits : {digitcount} Special = {specialcount}")
        
```

### 5) Word Frequency Counter (ignore the case)

### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
s = input("Enter a String : ")
result = ""
count = 1

for i in range(len(s) - 1):
    if s[i] == s[i + 1]:
        count += 1
    else:
        result += s[i] + str(count)
        count = 1

result += s[-1] + str(count)

print(result)
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
str = input("Enter a String : ")
ans = ""
for i in str:
    if i.isupper():
        ans += i.lower()
    else:
        ans += i.upper()

print(ans)
```

### 12) Palindrome Checker (ignore the case)
- `Madam` → Palindrome
- `Level` → Palindrome
- `Python` → Not Palindrome

```python
str = input("Enter a String : ").lower()
start = 0
end = len(str) - 1
is_palindrome = True
while start < end:
    if str[start] != str[end]:
        is_palindrome = False
    start += 1
    end -= 1
if is_palindrome : print(f"{str} is palindrome")
else :  print(f"{str} is not palindrome")
```

### 13) Longest Word Finder
- `Python programming is interesting` → programming
- `I love coding` → coding
- `Data structures and algorithms` → structures

```python
str = input("Enter a String : ")
word = str.split()
Max = 0
Maxword = ""
for i in word:
    length = len(i)
    if length > Max:
        Max = length
        Maxword = i
        
print(Maxword)
```
