### 03) WAP to create Circle class with area and perimeter function to find area and perimeter of circle.

```python
import math 
class Circle:
    def __init__(self , radius):
        self.radius = radius
    def area(self):
        return math.pi * self.radius * self.radius
    def perimeter(self):
        return 2 * math.pi * self.radius
        
c1 = Circle(10)
print(f"Area of Circle : {c1.area()}")
print(f"Perimeter of Circle : {c1.perimeter()}")
```

### 04) Create a class for employees that includes attributes such as name, age, salary, and methods to update and display employee information.

```python
class Employee():
    def __init__(self , name , age , salary):
        self.name = name
        self.age = age
        self.salary = salary
    def display(self):
        print(self.name , self.age , self.salary)
    def update(self , name , age , salary):
        self.name = name
        self.age = age
        self.salary = salary
        self.display()
        
e1 = Employee("Meet" , 18 , 10000)
e1.display()
e1.update("Meet" , 19 , 15000)
```

### 05) Create a bank account class with methods to deposit, withdraw, and check balance.

```python
class Bank_Account:
    def GetAccountDetails(self , Account_No , Account_Balance):
        self.Account_No = Account_No
        self.Account_Balance = Account_Balance
    def DisplayAccountDetails(self):
        print(self.Account_No , self.Account_Balance , end="\n")
    def Diposite(self , amt):
        self.Account_Balance += amt
        self.DisplayAccountDetails()
    def Withdraw(self , amt):
        if(amt > self.Account_Balance):
            print("Insufficient balance!")
        else:
            self.Account_Balance -= amt
        self.DisplayAccountDetails()

        
Account_No = input("Enter Account no : ")
Account_Balance = float(input("Enter Account Balance : "))

b1 = Bank_Account()
b1.GetAccountDetails(Account_No , Account_Balance)
b1.DisplayAccountDetails()
b1.Diposite(1000)
b1.Withdraw(500)
```

### 09) Define Time class with hour and minute as data member. Also define addition method to add two time objects.

```python
class Time:
    def __init__(self, hour, minute):
        self.hour = hour
        self.minute = minute

    def add(self, t):
        h = self.hour + t.hour
        m = self.minute + t.minute

   
        return Time(h, m)


t1 = Time(2, 50)
t2 = Time(1, 30)

result = t1.add(t2)

print(result.hour, "hour", result.minute, "minute")
```
